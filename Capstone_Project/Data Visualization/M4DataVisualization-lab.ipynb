# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# Set SQL query
QUERY = """
SELECT ConvertedComp
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(10,6))
sns.histplot(data=df, 
             x='ConvertedComp',
             bins=40)
plt.title('Histogram of Respondants Annual Salary (in USD)')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# Set SQL query
QUERY = """
SELECT Age
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(8,6))
sns.boxplot(data=df,
            y='Age',
            width=.23)
plt.title('Boxplot of Respondant Age')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# Set SQL query
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df = pd.read_sql_query(QUERY, conn)
df.plot(x='Age', 
        y='WorkWeekHrs', 
        kind='scatter',
        figsize=(8,6))
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# Set SQL query
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(10,6))
sns.scatterplot(data = df,
                x='WorkWeekHrs',
                y='CodeRevHrs',
                size= 'Age',
                hue='Age',
                palette='magma'
                )
plt.title('Hours Worked per Week compared to Hours of Code Revision')
plt.xlabel('Hours Worked')
plt.ylabel('Hours of Code Revision')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# Set SQL query
QUERY = """
SELECT LanguageDesireNextYear, 
       COUNT(LanguageDesireNextYear) as language_total, 
       SUM(COUNT(LanguageDesireNextYear)) OVER() AS total_count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY language_total DESC
LIMIT 5;
"""
df = pd.read_sql_query(QUERY, conn)

df['Percentage'] = df['language_total']/df['total_count']*100

labels= df['LanguageDesireNextYear'].to_list()

plt.figure(figsize=(10,10))
plt.pie(df['language_total'],
        autopct='%1.1f%%')
plt.title('Top 5 Databases Respondants Wish to Learn Next Year')
plt.legend(labels)
plt.show()
```
